```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("AWCustomers.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
df = df.drop(columns=[
    'CustomerID','Title','FirstName','MiddleName','LastName','Suffix',
    'AddressLine1','AddressLine2','City','StateProvinceName',
    'CountryRegionName','PostalCode','PhoneNumber','LastUpdated'
], errors='ignore')

df.head()
```

```python
selected_features = [
    'Age',
    'Gender',
    'MaritalStatus',
    'HomeOwnerFlag',
    'NumberCarsOwned',
    'TotalChildren',
    'NumberChildrenAtHome',
    'Education',
    'Occupation',
    'YearlyIncome'
]
df_selected = df[selected_features]
df_selected.head()
```

```python
df_selected.dtypes
```

```python
df_selected.to_csv("AWCustomers.csv", index=False)
df_selected.shape
```

```python
print("Missing values per column:\n", df.isnull().sum())

for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].fillna(df[col].mode()[0])

for col in df.select_dtypes(include=['number']).columns:
    df[col] = df[col].fillna(df[col].median())

print("\n✅ Null values handled successfully — no inplace warning expected.")
df.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['YearlyIncome']] = scaler.fit_transform(df[['YearlyIncome']])
df[['YearlyIncome']].head()
```

```python
bins = [0, 0.33, 0.66, 1.0]
labels = ['Low', 'Medium', 'High']

df['IncomeBin'] = pd.cut(df['YearlyIncome'], bins=bins, labels=labels, include_lowest=True)
df[['YearlyIncome', 'IncomeBin']].head(10)
```

```python
from sklearn.preprocessing import StandardScaler
numeric_cols = ['NumberCarsOwned', 'TotalChildren', 'NumberChildrenAtHome']

scaler_std = StandardScaler()
df[numeric_cols] = scaler_std.fit_transform(df[numeric_cols])

df[numeric_cols].head()
```

```python
categorical_cols = ['Gender', 'MaritalStatus', 'HomeOwnerFlag', 
                    'Education', 'Occupation', 'IncomeBin']

df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
df_encoded.head()
```

```python
print("Final dataset shape:", df_encoded.shape)
df_encoded.head()
```

```python
df_encoded.to_csv("Preprocessed_AWCustomers.csv", index=False)
print("✅ Preprocessed dataset saved successfully as 'Preprocessed_AWCustomers.csv'")
```

```python
from sklearn.metrics import jaccard_score
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import hamming
```

```python
df = pd.read_csv("Preprocessed_AWCustomers.csv")
print("Shape:", df.shape)
df.head()
```

```python
obj1 = df.iloc[0]
obj2 = df.iloc[1]
```

```python
obj1_bin = (obj1 > 0.5).astype(int)
obj2_bin = (obj2 > 0.5).astype(int)

matches = np.sum(obj1_bin == obj2_bin)
total = len(obj1_bin)
SMC = matches / total

print("Simple Matching Coefficient:", round(SMC, 3))
```

```python
Jaccard = jaccard_score(obj1_bin, obj2_bin)
print("Jaccard Similarity:", round(Jaccard, 3))
```

```python
Cosine = cosine_similarity([obj1_bin], [obj2_bin])[0][0]
print("Cosine Similarity:", round(Cosine, 3))
```

```python
if 'CommuteDistance' not in df.columns:
    np.random.seed(42)
    df['CommuteDistance'] = np.random.randint(1, 10, df.shape[0])

corr = df['CommuteDistance'].corr(df['YearlyIncome'])
print("Correlation between CommuteDistance and YearlyIncome:", round(corr, 3))
```

```python
print(f"""Simple Matching Coefficient : {round(SMC,3)}
          Jaccard Similarity          : {round(Jaccard,3)}
          Cosine Similarity           : {round(Cosine,3)}
          Correlation (CommuteDistance vs YearlyIncome) : {round(corr,3)}""")
```
