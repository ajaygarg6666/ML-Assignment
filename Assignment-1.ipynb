```python
import numpy as np
```

```python
# Q1. (a)
arr = np.array([1,2,3,6,4,5])
print("Original array:", arr)
arr_rev = arr[::-1]
print("Reversed array:", arr_rev)
```

```python
# Q1. (b)
array1 = np.array([[1, 2, 3], [2,4,5], [1,2,3]])
print("Original 2D array:\n", array1)
array1_flat = array1.flatten()
print("Flattened array:", array1_flat)
```

```python
# Q1. (c)
arr1 = np.array([[1, 2],[3, 4]])
arr2 = np.array([[1,2],[3,4]])
print("Array 1:\n", arr1)
print("Array 2:\n", arr2)
arr_compare = np.array_equal(arr1, arr2)
print("Are both arrays equal?", arr_compare)
```

```python
# Q1. (d)
x = np.array([1,2,3,4,5,1,2,1,1,1])
x_most_frequent = np.bincount(x).argmax()
print("Most frequent value in the array x :", x_most_frequent)
x_indices = np.where(x == x_most_frequent)[0]
print("Indices of the most frequent value:", x_indices)

y = np.array([1,1,1,2,3,4,2,4,3,3,])
y_most_frequent = np.bincount(y).argmax()
print("Most frequent value in the array y :", y_most_frequent)
y_indices = np.where(y == y_most_frequent)[0]
print("Indices of the most frequent value:", y_indices)
```

```python
# Q1. (e)
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print("Original array:", gfg)
gfg_sum = np.sum(gfg)
print("Sum of all elements in the array:", gfg_sum)
gfg_sumrowwise = np.sum(gfg, axis=1)
print("Sum of elements row-wise:", gfg_sumrowwise)
gfg_sumcolwise = np.sum(gfg, axis=0)
print("Sum of elements column-wise:", gfg_sumcolwise)
```

```python
# Q1. (f)
n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]])
n_array_sumdiagonal = np.trace(n_array)
print("Sum of diagonal elements:", n_array_sumdiagonal)
n_array_eigenvalue = np.linalg.eigvals(n_array)
print("Eigenvalues of the array:", n_array_eigenvalue)
n_array_eigenvector = np.linalg.eig(n_array)
print("Eigenvectors of the array:\n", n_array_eigenvector[1])
n_array_inverse = np.linalg.inv(n_array)
print("Inverse of the array:\n", n_array_inverse)
n_array_determinant = np.linalg.det(n_array)
print("Determinant of the array:", n_array_determinant)
```

```python
# Q1. (g)
p = [[1, 2], [2, 3]]
q = [[4, 5], [6, 7]]
multiplication_result = np.dot(p, q)
print("Result of matrix multiplication:\n", multiplication_result)
covariance_matrix = np.cov(p)
print("Covariance matrix of p:\n", covariance_matrix)

p = [[1, 2], [2, 3], [4, 5]]
q = [[4, 5, 1], [6, 7, 2]]
multiplication_result = np.dot(p, q)
print("Result of matrix multiplication:\n", multiplication_result)
covariance_matrix = np.cov(p)
print("Covariance matrix of p:\n", covariance_matrix)
```

```python
# Q1. (h)
x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])
inner_product = np.inner(x, y)
print("Inner product of x and y:\n", inner_product)
outer_product = np.outer(x, y)
print("Outer product of x and y:\n", outer_product)
cartesian_product = np.array(np.meshgrid(x, y)).T.reshape(-1, x.shape[1] + y.shape[1])
print("Cartesian product of x and y:\n", cartesian_product)
```

```python
# Q2. (a)
array = np.array([[1, -2, 3],[-4, 5, -6]])
absolute_array = np.abs(array)
print("Absolute array:\n", absolute_array)

flatten_percentiles = np.percentile(array.flatten(), [25, 50, 75])
print("25th, 50th, and 75th percentiles of the flattened array:", flatten_percentiles)

mean_value = np.mean(array)
print("Mean value of the array:", mean_value)
median_value = np.median(array)
print("Median value of the array:", median_value)
std_deviation = np.std(array)
print("Standard deviation of the array:", std_deviation)
```

```python
# Q2. (b)
a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0])
a_floor = np.floor(a)
print("Floor values of a:", a_floor)
a_ceil = np.ceil(a)
print("Ceil values of a:", a_ceil)
a_truncate = np.trunc(a)
print("Truncated values of a:", a_truncate)
a_rounded = np.round(a)
print("Rounded values of a:", a_rounded)
```

```python
# Q3. (a)
array = np.array([10, 52, 62, 16, 16, 54, 453])
array_sorted = np.sort(array)
print("Sorted array:", array_sorted)

array_indicesofsortedarray = np.argsort(array)
print("Indices of the sorted array:", array_indicesofsortedarray)

array_4smallestelements = np.partition(array, 4)[:4]
print("4 smallest elements in the array:", array_4smallestelements)

array_5largestelements = np.partition(array, -5)[-5:]
print("5 largest elements in the array:", array_5largestelements)
```
